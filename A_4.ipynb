```python
# Generate 1000 random numbers from normal distribution
mean = 0
std_dev = 1

data = np.random.normal(mean, std_dev, 1000)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black', density=True, alpha=0.6, color='skyblue')

# Plot theoretical normal curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)

plt.plot(x, p, 'r', linewidth=2)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
```

```python
# Simulate 1000 coin tosses
tosses = np.random.choice(['H', 'T'], size=1000)

heads = np.sum(tosses == 'H')
tails = np.sum(tosses == 'T')

print("Number of Heads:", heads)
print("Number of Tails:", tails)
print("Probability of Heads:", heads/1000)
print("Probability of Tails:", tails/1000)
```

```python
plt.bar(['Heads', 'Tails'], [heads, tails], color=['blue', 'red'])
plt.title("Coin Toss Simulation")
plt.ylabel("Count")
plt.show()
```
